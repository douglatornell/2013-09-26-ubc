##Shell Script Basics

A shell script is basically just a file containing a collections of commands that you
could have typed at the command line one after the other.
`bash` has some of the features of a programming language like
variables, loops, and conditionals,
but the syntax is quite different from Python's.

A simple `bash` script:

```bash
#!/usr/bin/env bash

# Hello bash

echo hello world!
echo from bash
echo
echo here is the contents of your directory:
ls
```

The first line is a special command called a "shebang"
(In the mists of time of Unnix history `#` was called "sharp" and `!` was called "bang").
It tells the operating system to use the program `/usr/bin/env` for find `bash`
and use it to run the commands in the file.

Comment lines in `bash` start with `#`.
That's why,
if your operating system doesn't know what to do with the shebang command on the 1st line
it ignores it.

The rest of the commands do exactly what they would do if you typed them in order.

###Running Shell Scripts and Permissions

If you save the commands above in a file
(let's call it `simple.sh`)
you can execute the script at the command line with:
```bash
$ source simple.sh
```
or

```bash
$ . simple.sh
```

`.` and `source` are synonyms.

Using `ls -l simple.sh` you will probably see that its permissions are something like:

```bash
-rw-r--r--  1 doug  staff     31 24 Sep 11:30 simple.sh
```

You can change the permissions to make the script executable with:

```bash
$ chmod +x simple.sh
$ ls -l simple.sh
-rwxr-xr-x  1 doug  staff     31 24 Sep 11:30 simple.sh
```

and,
having done so,
run the script with:

```bash
$ ./simple.sh
```

It's a security feature of `bash` that your current directory is not included in `PATH`,
so,
you have to prefix the script name with `./` to tell bash to run it from the current
directory.
If you put a shell script in a directory that is included in your `PATH` you can run
it by name,
just like any other command.
By convention,
people often put general purpose scripts that they have made executable for frequent use
in `$HOME/bin/` or `~/bin/` and add that directory to their `PATH` 
in their `$HOME/.bashrc` file with:

```bash
export PATH=$HOME/bin:$PATH
```

##Shell Variables and Loops

Variables are assigned in `bash` using `=` 
and their values are obtained by pre-fixing their names with `$`:

```bash
#!/usr/bin/env bash

FOO=bar
echo $FOO

RAZZLE="dazzle do"
echo $RAZZLE
```

Note that spaces are not allowed around `=`
and values that contain spaces must be quoted.

The `for` loop in `bash` allows us to iterate over a collection of argument values.
The arguments are given in a white-space separated list
and the loop syntax it:

```bash
#!/usr/bin/env bash

START_DATES="2013-08 2013-09"
DATA_FREQ="hourly"

for START_DATE in $START_DATES
do
    python climate_data.py $START_DATE $DATA_FREQ
done
```

If we decided instead that we wanted to store the command line argument pairs
for `climate_data.py` in a file
(let's call it `data_wanted.txt`),
on pair per line like:

    2010-08 hourly
    2010 daily

we can iterate over the lines read from `stdin` with the script:

```bash
#!/usr/bin/env bash

while read LINE
do
    python climate_data.py $LINE
done
```

and run it by piping our data file into our script:

```bash
$ cat data_wanted.txt | ./get_climate_data.sh
```

##Deferred Execution with `at`

Both Unix-like and Windows operating systems have an `at` command that allows
a shell script to be run at a future time.

The syntax differs significantly between Unix and Windows.

On OS/X and Linux the command:

```bash
at -f ./foo.sh 6pm tomorrow
```

will run the `foo.sh` script at 18:00 the next day,
and:

```bash
at -f ./foo.sh now + 3 hours
```

will run the script 3 hours in the future.

##Scheduled Execution with `cron`

`cron` is a Unix utility that allows jobs to be scheduled for repeated execution.
The schedule and commands to run are stored in a `crontab` file and accessed via the `crontab` command.
`man 5 crontab` provides all the details of the syntax that may be used in a `crontab` file:
Here is an example of a `crontab` that will run one script at midnight every day,
and another at 01:15 on the first day of January, April, July, and October:

    @midnight /path/to/daily_job.sh
    15 1 1 1,4,7,10 * /path/to/quarterly_job.sh
    
`cron` jobs are run as though you are working at the root (`/`) of the file system,
so you need to use absolute paths or include `cd` commands in your scripts.

The `/next` and `/every` flags of the Windows `at` command offer some `cron`-like functionality.

The same caveats mentioned above apply to scheduled jobs -
the computer must be away to execute them.

##Key Points
- Use `source` or `.` to run a shell script,
or make it executable so that you can invoke it with its path and name
- Shell variables are assigned like `FOO=bar` and used like `echo $FOO`
- `bash` provides a `for` loop construct to iterate over a list of values,
and a `while` loop to iterate over data from `stdin`
- On Linux and OS/X `cron` allows jobs to be run at scheduled times,
possibly on a repeating schedule
- On Windows the `/next` and `/every` flags of the `at` command provide some `cron`-like functions
- A computer must be awake to execute a deferred or scheduled job
