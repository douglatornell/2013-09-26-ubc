##`if __name__ == '__main__':`

`__name__` (pronounced "dunder name") is a special variable name that Python uses
to keep track of module's name.
When a module is imported its `__name__` is its file name
(without the `.py` extension).
When we tell Python to run a module at the shell prompt:
```bash
$ python climate_data.py
```
Python sets `__name__` to the string `__main__`.
That allows us to write some code that will only be executed when the module is
run at the command line and not when the module is imported.

By convention,
the `if __name__ == '__main__':` "stanza" is put at the end of a Python module.

We're going to use `if __name__ == '__main__':` to turn our `climate_data.py` module
into one that can be used at the command line to download an Environment Canada
Climate Archive file,
or in notebooks and scripts to import functions from.

What is the value of `sys.argv` each time the script is run?

by convention,
the amount of code in the `if __name__ == '__main__':` stanza if kept to a minimum;
usually just a function call to `main()` or `run()`.

So,
change your `if __name__ == '__main__':` stanza to:
```python
if __name__ == '__main__':
    main()
```
and add a `main()` function to your module:
```python
def main():
    assert len(sys.argv) == 2, \
        'Expected a year and month (yyyy-mm)'
    month = sys.argv[1]
    print 'Imma gonna go get the', month, 'data for Point Atkinson. Yo!'
```

##Parse the Command Line Arguments

We've made some promises in that docstring.
We say that we're going to handle 2 different configurations of
command line arguments:

1. `yyyy-mm hourly` to get a month's hourly climate data
2. `yyyy daily` to get a year's daily climate data

Some poeple call this "Documentation Driven Design".
The idea is that you right the documentation that describes
what your code is going to do,
and what its interfaces are,
and they you implement code to deliver on the promises in the docs.

So,
let's start delivering on those promises...

We'll start by noting that we're only interested in the 2nd and 3rd
command line arguments,
so we'll pass `sys.argv[1:]` into `main()`.
Our `if __name__ == '__main___':` clause becomes:
```python
if __name__ == '__main__':
    main(sys.argv[1:])
```
and we'll change our `main()` function to:
```python
def main(args):
    """Command line interface that downloads and stores a climate data file.
    """
    start_date, data_freq = parse_args(args)
    print 'Imma gonna go get the', start_date, data_freq, 'data for Point Atkinson. Yo!'
```

Creating a `parse_args()` function to parse the command line arguments
allows us to focus on the details doing that now as we develop the function
but hides those details in the flow of the higher level task of
downloading and storing climate data files.

A couple of things to note in the implementation of `parse_args()`:

1. There are all sorts of opportunities for errors when people type commands and arguments
so we use lots of `assert` statements to detect and be informative about invalid inputs.
2. `sys.argv` and thud the slice of it that we have in `args` is a list of strings
so we have to do some data type transformations to eventually return `start_date`
as a Python `datetime.date` object.

```python
def parse_args(args):
    assert len(args) == 2, \
        'Expected a month or year (yyyy-mm or yyyy) ' \
        'and a data frequency (hourly or daily)'
    data_freq = args[1]
    assert data_freq in ('hourly', 'daily'), \
        'Data frequency must be either hourly or daily'
    date_parts = args[0].split('-')
    if len(date_parts) == 1:
        assert data_freq == 'daily', \
            'A month is required for hourly data'
        year = int(date_parts[0])
        month = 1
    else:
        year = int(date_parts[0])
        month = int(date_parts[1])
        assert month in range(1, 13), \
            'Month must be in the range 1 to 12'
    start_date = datetime.date(year, month, 1)
    return start_date, data_freq
```
For the moment the only way we have to test all those `assert` statements
is to run our code over and over again at the command line.
Later we'll learn how to write code to test this code
and how to automate the running of that test code.

##Get the Data and Write the File

To finish up we'll add some code to:

- Set up the URL and `params` dict for a request to
the Environment Canada Climate Archive site for the type and period of data
that we specify in the command line arguments
- Calculate the file name that we'll write the data to
- Use `requests` to get the data
- Write the data to the file
- `print` a message to say what file the data were written to

Not surprisingly,
we're going to break those tasks up into a couple more functions:

```python
def main(args):
    """Command line interface that downloads and stores a climate data file.
    """
    start_date, data_freq = parse_args(args)
    url = 'http://climate.weather.gc.ca/climateData/bulkdata_e.html'
    params = calc_params(start_date, data_freq)
    filename = calc_filename(start_date, data_freq)
    response = requests.get(url, params=params)
    with open(filename, 'wb') as f:
        f.write(response.content)
    print data_freq, 'data for Point Atkinson for', start_date, 'written to', filename
```
```python
def calc_params(start_date, data_freq):
    """Calculate the params dict for the data request query.
    """
    params = {
        'stationID': 844,
        'format': 'csv',
    }
    params['Year'] = start_date.year
    params['Month'] = start_date.month
    params['Day'] = start_date.day
    if data_freq == 'hourly':
        params['timeframe'] = 1
    else:
        params['timeframe'] = 2
    return params
```
```python
def calc_filename(start_date, data_freq):
    """Calculate the filename based on start_date and data_freq.
    """
    filename = '-'.join((data_freq, 'PointAtkinson', str(start_date.year)))
    if data_freq == 'hourly':
        filename = '-'.join((filename, str(start_date.month)))
    filename = '.'.join((filename, 'csv'))
    return filename
```
We also need to `import requests` at the top of our module.

##Key Points
- An `if __name__ == '__main__':` stanza in a Python module isolates code
that will be run when the module is run via `python` at the command line
- `sys.argv` is a list of strings of the command line arguments,
starting with the file name of the Python module being run
- Module docstrings provide module-level documentation analogous to function docstrings
- Encapsulate details of different aspects of code in different functions
